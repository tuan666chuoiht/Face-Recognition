```python
import cv2
import pickle
import os
import numpy as np
```

```python
cam = cv2.VideoCapture(0)
#cam.set(3, 640)
#cam.set(4, 480)
```

```python
img_background = cv2.imread('./src/imagebackground.png')
img_background_tmp = cv2.imread('./src/imagebackground.png')
```

```python
# Khởi tạo bộ phát hiện khuôn mặt
faceDetect=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read('trainner.yml')
```

```python
# Mở file dê đọc
with open('./src/id_dict.pkl', 'rb') as file:
    id_dict = pickle.load(file)
    
print(id_dict)
    
def get_name_from_id(ids):
    name = id_dict[ids]
    return name
    
def getAvatar(ids):
    name_image = os.listdir('./src/PeopleID/' + str(ids) + '/' + 'avatar')[0]
    path = os.path.join('./src/PeopleID/', str(ids) + '/' + 'avatar' + '/' + str(name_image))
    avatar = cv2.imread(path)
    avatar = cv2.resize(avatar, (370, 370))
    return avatar
```

```python
#set text style
fontface = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
fontcolor = (0,255,0)
fontcolor1 = (0,0,255)
```

```python
i = 0
while True:
    i += 1
    success, frame = cam.read()
    frame_train = frame.copy()
    frame = cv2.resize(frame, (640,480))
    
    if not success:
        break
        
    gray=cv2.cvtColor(frame_train, cv2.COLOR_BGR2GRAY)

    faces = faceDetect.detectMultiScale(gray, 1.3, 5);
    
    img_background[120:120+480, 94:94+640] = frame
    
    img_tmp = img_background_tmp[470:470+90, 875:1210]
    
    for (x, y, w, h) in faces:
        x_resize = x + 94
        y_resize = y + 120
        cv2.rectangle(img_background, (x_resize, y_resize), (x_resize + w, y_resize + h), (255, 0, 0), 2)
        # Dự Doán
        ids, dist = recognizer.predict(gray[y:y + h, x:x + w])
        img_background[470:470+90, 875:1210] = img_tmp
    
        if dist <= 40:
            cv2.putText(img_background, str(get_name_from_id(str(ids))), (900,520), fontface, fontscale, fontcolor ,2)
            img_background[57:57+370, 860:860+370] = getAvatar(ids)
            print(dist)
            print(ids)
            print(get_name_from_id(str(ids)))
        else:
            cv2.putText(img_background, "Name: Unknown", (900,520), fontface, fontscale, fontcolor1, 2)
            unknow_avatar = cv2.imread('./src/unknow.jpg')
            unknow_avatar = cv2.resize(unknow_avatar, (370, 370))
            img_background[57:57+370, 860:860+370] = unknow_avatar
            print(dist)
            print(ids)
            print(get_name_from_id(str(ids)))
            
    cv2.imshow('Backround',img_background)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
cam.release()
cv2.destroyAllWindows()
```
