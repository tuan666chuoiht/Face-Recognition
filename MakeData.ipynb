```python
import cv2
import os
import random
import pickle
```

```python
cap = cv2.VideoCapture(0)
detector=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Mở file dê đọc
if not os.path.exists('./src/id_dict.pkl'):
    id_dict = {}
    with open('./src/id_dict.pkl', 'wb') as file:
        pickle.dump(id_dict, file)
else:
    with open('./src/id_dict.pkl', 'rb') as file:
        id_dict = pickle.load(file)
```

```python
print(id_dict)
```

```python
# Tạo Tên và ID
def generate_id():
    while True:
        # Tạo ID ngẫu nhiên
        random_id = random.randint(100000, 999999)
        # Kiểm tra xem ID đã tồn tại trong danh sách hay chưa
        if random_id not in id_dict.keys():
            # Nếu chưa, thêm ID mới vào danh sách và trả về ID
            return random_id
            break

name = 'Khanh Tuan'
random_id = generate_id()
id_dict[str(random_id)] = name
label = f"{random_id}"
```

```python
print(id_dict)
print(label)
print('./src/PeopleID/' + str(label))
```

```python
with open('./src/id_dict.pkl', 'wb') as file:
    pickle.dump(id_dict, file)
```

```python
i=0

while(True):
    i+=1
    success, frame = cap.read()
    if not success:
        continue
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #vẽ hình chữ nhật xunh quanh mặt
    faces = detector.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        # Vẽ hình chữ nhật quanh mặt nhận được
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        if i>=60:
            print("Số ảnh capture = ",i-60)
        # Tạo thư mục nếu chưa có
            if not os.path.exists('./src/PeopleID/' + str(label)):
                os.mkdir('./src/PeopleID/' + str(label))
            elif not os.path.exists('./src/PeopleID/' + str(label) + '/' + 'avatar'):
                os.mkdir('./src/PeopleID/' + str(label) + '/' + 'avatar')
        #lưu Dữ liệu
            path = os.path.normpath('./src/PeopleID/' + str(label) + "/" + str(i) + ".png")
            cv2.imwrite(os.path.join(path), gray[y:y + h, x:x + w])
    
    # Hiển thị
    cv2.imshow('frame',frame)

    if cv2.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
